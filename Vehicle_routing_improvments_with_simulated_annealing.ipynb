```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import random

# Define the number of customers and vehicles
n_customers = 100
n_vehicles = 3

# Define the depot location
depot = (18.19,  6.32)

# Generate random customer locations
np.random.seed(42) # for reproducibility
#customers = np.random.randint(-10, 10, size=(n_customers, 2))
customers=arr
# Define the distance function
def euclidean_distance(a, b):
    return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

# Define the initial solution as a list of routes
# Each route is a list of customer indices starting and ending with -1 (depot)
solution = [[-1] + random.sample(range(n_customers), k) + [-1] for k in np.random.multinomial(n_customers, [1/n_vehicles]*n_vehicles)]

# Define the objective function as the total distance of all routes
def objective(solution):
    total_distance = 0
    for route in solution:
        for i in range(len(route) - 1):
            a = depot if route[i] == -1 else customers[route[i]]
            b = depot if route[i+1] == -1 else customers[route[i+1]]
            total_distance += euclidean_distance(a, b)
    return total_distance

# Define a function to plot the solution
def plot_solution(solution):
    plt.figure()
    plt.scatter(customers[:, 0], customers[:, 1], color='b', label='Customers')
    plt.scatter(depot[0], depot[1], color='r', label='Depot')
    for i, route in enumerate(solution):
        color = np.random.rand(3,)
        for j in range(len(route) - 1):
            a = depot if route[j] == -1 else customers[route[j]]
            b = depot if route[j+1] == -1 else customers[route[j+1]]
            plt.plot([a[0], b[0]], [a[1], b[1]], color=color, label=f'Vehicle {i+1}' if j == 0 else '')
    plt.legend(loc='upper left')
    plt.title(f'Total distance: {objective(solution):.2f}')
    plt.show()

# Plot the initial solution
plot_solution(solution)

# Define a function to generate a random neighbor solution by swapping two customers in different routes
def swap_neighbor(solution):
    # Copy the solution as a list of lists
    neighbor = [route.copy() for route in solution]
    # Choose two random routes
    i, j = np.random.choice(len(neighbor), size=2, replace=False)
    # Choose two random customers from the routes
    k, l = np.random.choice(len(neighbor[i]) - 2, size=2) + 1 # exclude depot (-1)
    m, n = np.random.choice(len(neighbor[j]) - 2, size=2) + 1 # exclude depot (-1)
    # Swap the customers
    neighbor[i][k], neighbor[j][m] = neighbor[j][m], neighbor[i][k]
    return neighbor

# Define the metaheuristic algorithm (Simulated Annealing)
def simulated_annealing(solution, max_iter, init_temp, alpha):
    # Initialize the current solution and its objective value
    current = solution
    current_obj = objective(current)
    # Initialize the best solution and its objective value
    best = current
    best_obj = current_obj
    # Initialize the temperature and the iteration counter
    temp = init_temp
    iter = 0
    # Loop until the maximum number of iterations is reached
    while iter < max_iter:
        # Generate a random neighbor solution
        neighbor = swap_neighbor(current)
        # Calculate its objective value
        neighbor_obj = objective(neighbor)
        # Calculate the difference with the current objective value
        diff = neighbor_obj - current_obj
        # If the neighbor is better than the current solution, accept it
        if diff < 0:
            current = neighbor
            current_obj = neighbor_obj
            # If the neighbor is better than the best solution, update it
            if neighbor_obj < best_obj:
                best = neighbor
                best_obj = neighbor_obj
        # If the neighbor is worse than the current solution, accept it with a probability based on the temperature
        else:
            prob = np.exp(-diff / temp)
            if np.random.rand() < prob:
                current = neighbor
                current_obj = neighbor_obj
        # Decrease the temperature by a factor of alpha
        temp = temp * alpha
        # Increase the iteration counter by one
        iter = iter + 1
    # Return the best solution and its objective value
    return best, best_obj

# Set the parameters of the algorithm
max_iter = 1000 # maximum number of iterations
init_temp = 10 # initial temperature
alpha = 0.99 # cooling factor

# Run the algorithm and print the results
best, best_obj = simulated_annealing(solution, max_iter, init_temp, alpha)
print(f'Best solution: {best}')
print(f'Best objective value: {best_obj:.2f}')

# Plot the best solution
plot_solution(best)
```
